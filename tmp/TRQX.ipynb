```python
%%time

iO = parser.litOrders[parser.litOrders['EXECTYPE'] == 'Insert'][['SIDE', 'PRICE', 'VISIBLEQTY']]
_l1 = parser.get_l1(iO.index.values, post_clearing=True)
```

```python
def check(x):
    return x.index[0] == 0.0 and x.shape[0] == 1

def get_tick_size(x):
    bins = [.001, .002, .005, .01, .02, .05, .1][::-1]
    return bins[np.argmax([check(pd.Series(np.round(np.mod(np.round(prcs / k, 6), 1), 6))
                                 .value_counts().astype(int)) for k in bins])]
```

```python
prcs = parser.litOrders['PRICE'].values

tick_size = .02
__l1 = _l1.shift(1)
iO = parser.litOrders[parser.litOrders['EXECTYPE'] == 'Insert'][['SIDE', 'PRICE', 'VISIBLEQTY']].copy()
iO['SIDE'] = (iO['SIDE'] == 'Sell').astype(float)
iO['bb'] = __l1['best_bid_prc']
iO['ba'] = __l1['best_ask_prc']
iO['dp'] = (1. - iO['SIDE']) * (iO['bb'] - iO['PRICE']) + iO['SIDE'] * (iO['PRICE'] - iO['ba'])
iO['dpx'] = (1. - iO['SIDE']) * (iO['ba'] - iO['PRICE']) + iO['SIDE'] * (iO['PRICE'] - iO['bb'])
iO['spread'] = iO['ba'] - iO['bb']
iO[['PRICE', 'bb', 'ba', 'dp', 'dpx', 'spread']] = np.round(iO[
    ['PRICE', 'bb', 'ba', 'dp', 'dpx', 'spread']] / tick_size)
iO
```

```python
iO['dpx'].value_counts().sort_index().loc[-10:100].plot()
iO['dp'].value_counts().sort_index().loc[-10:100].plot()
```
