```python
# function to optimize; maps ABM model parameters to model-real distances
def func_opt(zi_mean, n_sigma, f_kappa, f_kappa_3, lf_beta, hf_beta, mm_lo, opt_params):
    tmp = {}
    tmp['zi_mean'], tmp['n_sigma'], tmp['f_kappa'], tmp['f_kappa_3'], tmp['lf_beta'], tmp['hf_beta'], tmp['mm_lo'] = zi_mean, n_sigma, f_kappa, f_kappa_3, lf_beta, hf_beta, mm_lo
    update_model_params(model_params, tmp)
    save_target_params(model_params, save_path='CME/model_params.json')
    flag = run_simulation()
    assert flag == 0, "Error when running simulation!"
    names = [x for x in os.listdir(opt_params['res_directory']) if 'run_' in x]
    tmp_dist = []
    for n_ in names:
        data_path = opt_params['res_directory'] + n_
        sim_md = get_synthetic_market_data_sim(data_path)
        sim_ret = get_simulated_ret(sim_md, opt_params['start'], opt_params['end'], freq=opt_params['freq'])
        dist = prepare_distance(opt_params['his_ret'], sim_ret, opt_params['LAGS'], opt_params['LAGS2'])
        tmp_dist.append(dist)
    print(tmp_dist)
    intermediate = (np.array(tmp_dist).mean(axis=0) + np.array(tmp_dist).std(axis=0)) * opt_params['weights']
    return max(-1. * intermediate.sum(), np.float64(-1))

def func_opt_moment(zi_mean, n_sigma, f_kappa, f_kappa_3, lf_beta, hf_beta, mm_lo, opt_params):
    tmp = {}
    tmp['zi_mean'], tmp['n_sigma'], tmp['f_kappa'], tmp['f_kappa_3'], tmp['lf_beta'], tmp['hf_beta'], tmp['mm_lo'] = zi_mean, n_sigma, f_kappa, f_kappa_3, lf_beta, hf_beta, mm_lo
    update_model_params(model_params, tmp)
    save_target_params(model_params, save_path='CME/model_params.json')
    flag = run_simulation()
    assert flag == 0, "Error when running simulation!"
    names = [x for x in os.listdir(opt_params['res_directory']) if 'run_' in x]
    moment_cover = np.array([0, 0] + [0 for _ in range(len(opt_params['LAGS'])//opt_params['ll'])] + [0 for _ in range(len(opt_params['LAGS2'])//opt_params['ll'])] + [0])
    for n_ in names:
        data_path = opt_params['res_directory'] + n_
        sim_md = get_synthetic_market_data_sim(data_path)
        sim_ret = get_simulated_ret(sim_md, opt_params['start'], opt_params['end'], freq=opt_params['freq'])
        sim_cover = get_moment_coverage(sim_ret, opt_params['interval'], opt_params['ll'], opt_params['LAGS'], opt_params['LAGS2'])
        moment_cover += np.array(sim_cover)
    print(np.round(moment_cover / len(names), 4).tolist())
    ratio = np.min(moment_cover[:-1]) / len(names)
    return ratio + 0.01
```

```python
def xgboost_modelling(n_dims, total_num, params_range, initial_eval, opt_params, func_opt, batch_num=10, n_iters=20):
    total_samples = get_sobol_samples(n_dims, total_num, params_range)# Total sample space
    np.random.shuffle(total_samples)
    x_evaluated, x_remaining = split_samples(total_samples, number=initial_eval)
    y_evaluated = np.array([])
    for i, xx in enumerate(x_evaluated):
        if i%10==0:
            print(i)
        y_evaluated = np.append(y_evaluated, func_opt(xx[0], xx[1], xx[2], xx[3], xx[4], xx[5], xx[6], opt_params))
    
    tmp_ind = y_evaluated.argsort()[-5:] # Find the index for the best 5 points
    tmp_opt = x_evaluated[tmp_ind[-1]]
    print('Current best evaluated: ', tmp_opt, y_evaluated[tmp_ind[-1]], max(y_evaluated))
    # Fit surrogate using the initial chosen random points
    surrogate_model_XGBoost = fit_surrogate_model(x_evaluated, y_evaluated)
    # Iteratively choose more points and fit the model
    for k in range(n_iters):
        print(k)
        y_hat = surrogate_model_XGBoost.predict(x_remaining)
        # ind_eval = y_hat.argsort()[-batch_num:]
        ind_eval = np.concatenate([y_hat.argsort()[-batch_num:], np.random.choice(len(y_hat), size=batch_num//3, replace=False)])
        x_to_eval, x_remaining = split_samples(x_remaining, ind=ind_eval)
    
        y_to_eval = np.array([])
        for i, xx in enumerate(x_to_eval):
            y_evaluated = np.append(y_evaluated, func_opt(xx[0], xx[1], xx[2], xx[3], xx[4], xx[5], xx[6], opt_params))
        x_evaluated = np.concatenate([x_evaluated, x_to_eval])
        y_evaluated = np.concatenate([y_evaluated, y_to_eval])
        tmp_ind = y_evaluated.argsort()[-5:] # Find the index for the best 5 points
        tmp_opt = x_evaluated[tmp_ind[-1]]
        print('Current best evaluated: ', tmp_opt, y_evaluated[tmp_ind[-1]], max(y_evaluated))
        surrogate_model_XGBoost = fit_surrogate_model(x_evaluated, y_evaluated)
        
    
    total_y_hat = surrogate_model_XGBoost.predict(total_samples)
    print(total_y_hat.shape, total_y_hat.max())
    global_ind = total_y_hat.argsort()[-5:] # Find the index for the best 5 points
    opt = total_samples[global_ind[-1]] # Find the best point
    print('opt: ', opt)
    pred_y_opt = surrogate_model_XGBoost.predict(opt.reshape(1,-1))
    simulated_y_opt = func_opt(opt[0], opt[1], opt[2], opt[3], opt[4], opt[5], opt[6], opt_params)
    print('predicted optimal: %.6f; Actual simulated value in the predicted optimal point: %.6f'%(pred_y_opt, simulated_y_opt))
    return opt, surrogate_model_XGBoost, pred_y_opt, simulated_y_opt
```
